2. Peça um número inteiro e informe se ele é par. Crie um algoritmo em Python que verifique se o número é divisível por 2.


```python
# Solicita um número inteiro ao usuário
numero_str = input("Digite um número inteiro: ")

# Converte a entrada para um número inteiro
try:
    numero = int(numero_str)

    # Verifica se o número é divisível por 2 usando o operador módulo (%)
    if numero % 2 == 0:
        print(f"O número {numero} é par.")
    else:
        print(f"O número {numero} é ímpar.")

except ValueError:
    print("Entrada inválida. Por favor, digite um número inteiro.")
```

10. Peça a quantidade de horas trabalhadas e o valor por hora. Se trabalhou até 40 horas, calcule o salário normalmente (horas × valor por hora). Se trabalhou mais de 40 horas, as horas extras devem ter um adicional de 50% (horas extras × valor por hora × 1.5). Crie um algoritmo em Python que realize esse cálculo corretamente.

```python
# Solicita os dados ao usuário
horas_trabalhadas = float(input("Digite a quantidade de horas trabalhadas: "))
valor_hora = float(input("Digite o valor por hora: "))

# Verifica se há horas extras
if horas_trabalhadas <= 40:
    salario = horas_trabalhadas * valor_hora
else:
    horas_extras = horas_trabalhadas - 40
    salario = (40 * valor_hora) + (horas_extras * valor_hora * 1.5)

# Exibe o resultado
print(f"Seu salário é R${salario:.2f}")
```

O objetivo é criar um programa Python que simule uma calculadora interativa, capaz de realizar diferentes tipos de análises e conversões, e que só termine quando o usuário explicitamente decidir sair.



```python
# 1. Estrutura Principal e Menu (Usando while)
while True:
    print("\n" + "="*25)
    print("**CALCULADORA INTERATIVA**")
    print("="*25)
    print("Digite o número da opção desejada:")
    print("1 - Múltiplos de 5 em um Intervalo")
    print("2 - Contagem de Números Pares em um Intervalo")
    print("3 - Verificador de Número Primo")
    print("4 - Conversor de Temperatura (Cº para Fº)")
    print("0 - Sair do Programa")
    print("-" * 25)

    # Solicita a opção e usa .strip() para remover espaços em branco acidentais
    # e .lower() (embora só tratemos números) para padronizar a entrada
    opcao = input("Opção: ").strip()

    # 2. Tratamento das Opções (Usando if/elif/else)

    if opcao == '0':
        # Se o usuário digitar 0: Interrompe o loop e exibe a mensagem de saída
        print("\n Saindo do programa. Até mais!")
        break

    elif opcao == '1':
        # Opção 1: Múltiplos de 5 em um Intervalo (Usando for)
        print("\n--- 1. Múltiplos de 5 em um Intervalo ---")
        try:
            # Pede o número inicial e final
            num_inicial = int(input("Digite o número inicial do intervalo: "))
            num_final = int(input("Digite o número final do intervalo: "))

            # Garante que o número inicial seja menor ou igual ao final para o loop funcionar corretamente
            if num_inicial > num_final:
                num_inicial, num_final = num_final, num_inicial # Troca os valores

            print(f"\n Múltiplos de 5 entre {num_inicial} e {num_final}:")

            # Variável para rastrear se algum múltiplo foi encontrado
            encontrou_multiplo = False

            # Loop for para percorrer o intervalo (o + 1 garante que o número final seja incluído)
            for numero in range(num_inicial, num_final + 1):
                # O operador de módulo (%) retorna o resto da divisão. Se for 0, é múltiplo de 5.
                if numero % 5 == 0:
                    print(numero)
                    encontrou_multiplo = True

            if not encontrou_multiplo:
                 print("Nenhum múltiplo de 5 encontrado no intervalo.")

        except ValueError:
            print(" Erro: Por favor, digite números inteiros válidos para o intervalo.")

    elif opcao == '2':
        # Opção 2: Contagem de Números Pares em um Intervalo (Usando for)
        print("\n--- 2. Contagem de Números Pares em um Intervalo ---")
        try:
            # Pede o número inicial e final
            num_inicial = int(input("Digite o número inicial do intervalo: "))
            num_final = int(input("Digite o número final do intervalo: "))

            if num_inicial > num_final:
                num_inicial, num_final = num_final, num_inicial # Troca os valores

            contador_pares = 0

            # Loop for para percorrer o intervalo
            for numero in range(num_inicial, num_final + 1):
                # Se o resto da divisão por 2 for 0, o número é par
                if numero % 2 == 0:
                    contador_pares += 1 # Incrementa o contador

            # Exibe o resultado da contagem
            print(f"\n Total de números pares entre {num_inicial} e {num_final}: **{contador_pares}**")

        except ValueError:
            print(" Erro: Por favor, digite números inteiros válidos para o intervalo.")

    elif opcao == '3':
        # Opção 3: Verificador de Número Primo (Usando for e if)
        print("\n--- 3. Verificador de Número Primo ---")
        try:
            # Pede o número inteiro positivo
            numero = int(input("Digite um número inteiro positivo: "))

            if numero <= 1:
                # 1 e números menores/iguais a 0 não são primos por definição
                print(f"\n O número {numero} **não é primo** (Primos devem ser maiores que 1).")
            else:
                # Assume que o número é primo inicialmente
                e_primo = True

                # Loop for para verificar divisibilidade
                # Começa em 2 porque todo número é divisível por 1
                # Vai até a raiz quadrada do número (int(numero**0.5) + 1) para otimizar,
                # mas seguiremos a sugestão mais didática de ir até numero - 1.
                # Para maior eficiência, a verificação vai de 2 até (numero // 2) + 1
                for i in range(2, numero):
                    if numero % i == 0:
                        # Se for divisível por qualquer número neste intervalo, não é primo
                        e_primo = False
                        break # Não precisa continuar verificando

                # Exibe o resultado
                if e_primo:
                    print(f"\n O número {numero} **é primo**.")
                else:
                    print(f"\n O número {numero} **não é primo**.")

        except ValueError:
            print(" Erro: Por favor, digite um número inteiro positivo válido.")

    elif opcao == '4':
        # Opção 4: Conversor de Temperatura (Cº para Fº) (Estrutura simples)
        print("\n--- 4. Conversor de Temperatura (Cº para Fº) ---")
        try:
            # Pede a temperatura em Celsius
            celsius = float(input("Digite a temperatura em Celsius (Cº): "))

            # Fórmula: F = C * 1.8 + 32
            fahrenheit = celsius * 1.8 + 32

            # Exibe o valor convertido
            # Usando f-string para formatar com 2 casas decimais
            print(f"\n {celsius:.2f}ºC é equivalente a **{fahrenheit:.2f}ºF**.")

        except ValueError:
            print(" Erro: Por favor, digite um valor numérico válido para a temperatura.")

    else:
        # Se o usuário digitar qualquer outra coisa: Opção inválida
        print("\n Opção inválida! Por favor, digite um número de 0 a 4.")
```
